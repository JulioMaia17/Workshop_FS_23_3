```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("dados_rh.csv")
display(df)
```

```python
df.shape
display("quantidade de linhas e colunas respectivamente e: {}".format(df.shape))
```

```python
dn = df.isnull().sum().sum()

if dn == 0:
  print("Sem dados nulos na base de dados")
else:
  print("Tem dados nulos na base de dados")
```

```python
td = df.dtypes
print("Tipos de dados presentes no banco de dados:")
print(td)
```

```python
tdu = df.dtypes.unique()

print("Os tipos de dados presente nesse banco de dados sao:")
for tipo in tdu:
    print(tipo)

qtd = len(tdu)

if qtd== 2:
  print("Tem dois tipos de dados")
elif qtd > 2:
  print("Tem mais de dois tipos de dados")
elif qtd == 1:
  print("Tem um tipo de dados")
else:
  print("Nao tem nenhum tipo de dados")
```

```python
quantidade = df.groupby('Department')['EmployeeNumber'].count()
```

```python
educacao = df.groupby('Education')['EmployeeNumber'].count()
```

```python
atrito= df.groupby('Attrition')[['EmployeeNumber']].sum()
```

```python

# Crie o gráfico de barras
quantidade.plot(kind='bar')
plt.title('Quantidade de vagas em cada setor')
plt.xlabel('Setores')
plt.ylabel('Quantidade de Funcionários')
display(quantidade)
```

```python
# Crie o gráfico de linha
educacao.plot(kind='line', marker='o')
plt.title('QUANTIDADE DE FUNCIONARIOS POR NIVEL DE ESCOLARIDADE')
plt.xlabel('ESCOLARIDADE')
plt.ylabel('QUANTIDADE DE FUNCIONARIOS ')
display(educacao)
```

```python
atrito.plot(kind = 'pie', subplots = True)
```
