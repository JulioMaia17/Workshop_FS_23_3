```python
import plotly.graph_objects as go

data = pd.read_csv('desmatamento_prodes.csv')

fig = go.Figure(data=[go.Table(
    header=dict(values=['Área Desmatada', 'Referência']),
    cells=dict(values=[data['area_total_desmatamento'], data['referencia']])
)])

fig.update_layout(
    title='Comparação de Área Desmatada e Referência',
)

fig.show()
```

```python

import plotly.graph_objects as go

data = pd.read_csv('desmatamento_prodes.csv')

fig = go.Figure(data=[
    go.Bar(x=data['referencia'], y=data['area_total_desmatamento'])
])

fig.update_layout(
    title='Área Desmatada por Ano',
    xaxis_title='Referência',
    yaxis_title='Área Desmatada'
)

fig.show()
```
